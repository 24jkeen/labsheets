# Do I need to read this?

Read through this if any of the following is *not* true

* `python` launches a python REPL (interpreter)
* `python --version` prints `2.7.X` or `python2 --version` prints `2.7.X`
* `pip -V` prints out the version of pip and lists its installation inside a `python2.7` directory (probably under `/usr/lib`
* `jupyter-notebook` starts a jupyter server running on http://localhost:8888

# Installing Python locally

## Linux

It is very likely that you will already have python installed out of the box,
however if you don't or wish to upgrade to a more recent version, one of the
following commands might help you.

### Ubuntu

```
$ sudo apt update && apt install python2.7 python-pip
```

### Fedora


```
$ sudo dnf install python27 python-pip
```

[Fedora developer
portal](https://developer.fedoraproject.org/tech/languages/python/multiple-pythons.html)
has some good docs on managing python versions.

### Arch

```
$ sudo pacman -Sy python2 python-pip
```

## Windows


Use [chocolatey](https://chocolatey.org/), or alternatively the [MSI installer](https://www.python.org/downloads/windows/)

```
C:\> choco install python2
```

## macOS

Python comes installed out of the box on OS X/macOS however you might find you
want to upgrade to a more recent version. [`brew`](https://brew.sh/) has you covered. 


```
$ brew install python
```

**NOTE**: Ensure you are using Python 2 and not Python 3, as BC4 is set up for Python 2.

```
$ python --version
Python 2.7.13
```

You should see a `2.7.x` version number.

# Installing Jupyter

Having installed python, you should also have access to `pip`, the standard python package manager.

Install Jupyter by running

```bash
$ pip install --user jupyter
```

The `--user` flag will install to to `~/.local` on a unix system and `%APPDATA%/Python` on a windows 
system. Ensure you have the bin folder on your path.

If you already have Jupyter, be sure to upgrade it with a 

```bash
$ pip install --user --upgrade jupyter
```

Finally you should be able to launch a Jupyter server by running, you should see an output similar 
to that below.
